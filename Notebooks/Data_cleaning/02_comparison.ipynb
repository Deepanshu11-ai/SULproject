```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_unclean = pd.read_csv("/Users/deepanshus/StockMarketPredictor/Project/data/Combined/all/unclean_data.csv")
df_clean = pd.read_csv("/Users/deepanshus/StockMarketPredictor/Project/data/Combined/all/final_processed_data.csv")
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
sns.heatmap(df_unclean.isnull(), cbar=False)
plt.title("Before Cleaning (Missing Values)")

plt.subplot(1,2,2)
sns.heatmap(df_clean.isnull(), cbar=False)
plt.title("After Cleaning (No Missing Values)")

plt.show()
```

```python
plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
sns.boxplot(data=df_unclean.select_dtypes(include='number'))
plt.title("Before Cleaning (Outliers Present)")
plt.xticks(rotation=45)

plt.subplot(1,2,2)
sns.boxplot(data=df_clean)
plt.title("After Cleaning (Outliers Removed)")
plt.xticks(rotation=45)

plt.show()
```

```python
feature = "volume_change"  

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
df_unclean[feature].hist(bins=30)
plt.title(f"{feature} Before Cleaning")

plt.subplot(1,2,2)
df_clean[feature].hist(bins=30)
plt.title(f"{feature} After Cleaning")

plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.scatter(df_unclean["rsi"], df_unclean["target"])
plt.title("Before Cleaning (Noisy)")
plt.xlabel("RSI")
plt.ylabel("Target")

plt.subplot(1,2,2)
plt.scatter(df_clean["rsi"], df_clean["target"])
plt.title("After Cleaning (Clearer Pattern)")
plt.xlabel("RSI")
plt.ylabel("Target")

plt.show()
```

```python
plt.figure(figsize=(14,5))

# BEFORE (unclean)
plt.subplot(1,2,1)
sns.heatmap(df_unclean.apply(pd.to_numeric, errors='coerce').corr(), cmap="coolwarm")
plt.title("Before Cleaning")

# AFTER (clean)
plt.subplot(1,2,2)
sns.heatmap(df_clean.corr(), cmap="coolwarm")
plt.title("After Cleaning")

plt.show()
```

```python
summary = pd.DataFrame({
    "Metric": ["Rows", "Missing Values"],
    "Before": [len(df_unclean), df_unclean.isnull().sum().sum()],
    "After": [len(df_clean), df_clean.isnull().sum().sum()]
})

print(summary)
```
